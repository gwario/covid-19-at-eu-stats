```python
import pandas
import data
import os

input_output_dir = data.data_dir

# The year to select from country stats
data_date = data.worldbank_data_column_date
# The output file containing only the selected country data
output_filename = data.worldbank_data_filename


# The column containing the name of the data of one file from the worldbank data sets
indicator_column_name = 'Indicator Name'

df_country_data = None

for entry in os.scandir(input_output_dir):
    
    in_filename = entry.name
    
    if in_filename != output_filename and in_filename.endswith(".csv") and not in_filename.startswith("Metadata") and entry.is_file():
        
        print("Loading {}".format(entry.name))
        
        df = pandas.read_csv(entry.path, index_col=0, skiprows=4)
        
        # Get the actual name of the indicator
        indicator_name = df.loc[:, indicator_column_name].values[0]
        
        # Correctly name our target column
        target_column = indicator_name+" in "+data_date
        df.rename(columns={data_date: target_column}, inplace=True)
        
        # Only keep target column and append it to the dataset
        if df_country_data is None:
            # Create dataframe
            df_country_data = pandas.DataFrame({target_column: df.loc[:, target_column]})
        else:
            # Append to dataframe
            df_country_data = pandas.concat([df_country_data, df.loc[:, target_column]], axis=1)

print("Loaded country data:")
print(df_country_data.columns.values)
print(df_country_data.index.values)

def rename(df):
    df.rename(index={'United Kingdom': data.uk,
                     'Syrian Arab Republic': data.sy,
                     'Slovak Republic': data.sk,
                     'United States': data.us,}, inplace=True)

# Fix country names to hopkins data
rename(df_country_data)

df_country_data.to_csv(input_output_dir+output_filename)
```
